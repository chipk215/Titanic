#### Read the test and training data files and combine for analysis. 
Add the survived column to the test data and set the value to -1 to indocate unknown.
The PassengerIds are unique in each file.

```python
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
test.insert(loc=1, column='Survived', value=-1)
data = pd.concat([train, test], ignore_index=True)
```

#### Metadata based upon https://www.kaggle.com/mgmarques/learning-ml-with-titanic 
Notes: 




```python
def get_meta_data(df, prediction=None):
    samples = df.shape[0]
    types = df.dtypes
    counts = df.apply(lambda x: x.count())
    uniques = df.apply(lambda x: [x.unique()])
    nulls = df.apply(lambda x: x.isnull().sum())
    distincts = df.apply(lambda x: x.unique().shape[0])
    missing_ratio = (df.isnull().sum()/ samples) * 100
    print('Data shape:', df.shape)
    if prediction is None:
        cols = ['types', 'counts', 'distincts', 'nulls', 'missing ratio', 'uniques']
        str = pd.concat([types, counts, distincts, nulls, missing_ratio, uniques], axis = 1)
    else:
        corr = df.corr()[prediction]
        str = pd.concat([types, counts, distincts, nulls, missing_ratio, uniques, corr], axis = 1, sort=False)
        corr_col = 'corr '  +  prediction
        cols = ['types', 'counts', 'distincts', 'nulls', 'missing_ratio', 'uniques', corr_col ]
        
    str.columns = cols
    dtypes = str.types.value_counts()
    print('___________________________\nData types:\n',str.types.value_counts())
    print('___________________________')
    return str
```

```python
details = get_meta_data(data.loc[:, 'Survived' : 'Embarked'], 'Survived')
details.sort_values(by='corr Survived', ascending=False)                                 
```

#### What is the surival rate for passengers traveling on a single ticket?

Reminder, the survival rate for all of the passengers in the training set is 38.38%.

#### What is the surival rate for passengers traveling on a group ticket?

```python
unique_training_tickets = train.Ticket.value_counts()

temp = unique_training_tickets.reset_index()
temp =temp.rename(columns={'index': 'TicketNumber','Ticket': 'PassengerCount'})
temp =temp.rename(columns={'TicketNumber': 'Ticket','Ticket': 'PassengerCount'})
single_ticket_passengers = train.loc[train['Ticket'].isin(temp[temp['PassengerCount']==1]['Ticket'])]
single_ticket_survivor_count = single_ticket_passengers.Survived.sum()
print("Single Ticket Passenger Count= ", single_ticket_passengers.shape[0])
print('Single Ticket Survivor Rate:', single_ticket_survivor_count/single_ticket_passengers.shape[0])

group_ticket_passengers = train.loc[~train['Ticket'].isin(temp[temp['PassengerCount']==1]['Ticket'])]
group_ticket_survivor_count = group_ticket_passengers.Survived.sum()
print("Group Ticket Passenger Count= ", group_ticket_passengers.shape[0])
print('Group Ticket Survivor Rate:', group_ticket_survivor_count/group_ticket_passengers.shape[0])
```

```python
def map_title(title):
    title_dictionary = {
        "Mr." : "Mr",
        "Miss." : "F_Unwed",
        "Mlle." : "F_Unwed",
        "Ms." : "F_Unwed",
        "Mrs." : "F_Wed",
        "Mme." : "F_Wed",
        "Master." : "Master",
        "Rev." : "Religon",
        "Dr." : "Professional",
        "Col." : "Professional",
        "Major." : "Professional",
        "Capt." : "Professional",
        "Lady." : "Royalty",
        "Sir." : "Royalty",
        "the" : "Royalty",
        "Jonkheer." : "Royalty",
        "Don." : "Royalty",
        "Dona." : "Royalty"   
    }
    
    return title_dictionary.get(title, "Other")
```
